```python
import pandas as pd
import numpy as np
```

```python
csp_accompanied = pd.read_csv('csp_accompanied.csv')
```

```python
dfa = csp_accompanied.set_index(['Unnamed: 0'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covt = np.cov(dfa.iloc[:60,:2])
sns.heatmap(covt, cbar_kws={"shrink": .5})
```

```python
subj1 = dfa[dfa.index.str.contains('subj1')]
```

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
```

```python
classif = OneVsRestClassifier(SVC(kernel='rbf'))
data = subj1.iloc[:,0:4]
target = subj1.iloc[:,4:]
classif.fit(data, target)
```

```python
classif.score(data, target)
```
